```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
CFM_FP = './data/time_series_covid19_confirmed_global.csv'
DTH_FP = './data/time_series_covid19_deaths_global.csv'
RCV_FP = './data/time_series_covid19_recovered_global.csv'
```

```python
cfm_df = pd.read_csv(CFM_FP)
dth_df = pd.read_csv(DTH_FP)
rcv_df = pd.read_csv(RCV_FP)
```

```python
cfm_df_india = cfm_df[cfm_df['Country/Region'] == 'India']
del cfm_df_india['Country/Region']
del cfm_df_india['Province/State']
del cfm_df_india['Lat']
del cfm_df_india['Long']

dth_df_india = dth_df[dth_df['Country/Region'] == 'India']
del dth_df_india['Country/Region']
del dth_df_india['Province/State']
del dth_df_india['Lat']
del dth_df_india['Long']

rcv_df_india = rcv_df[rcv_df['Country/Region'] == 'India']
del rcv_df_india['Country/Region']
del rcv_df_india['Province/State']
del rcv_df_india['Lat']
del rcv_df_india['Long']
```

```python
cfm_df_plot = cfm_df_india.transpose()
cfm_df_plot.columns = ['count']

dth_df_plot = dth_df_india.transpose()
dth_df_plot.columns = ['count']

rcv_df_plot = rcv_df_india.transpose()
rcv_df_plot.columns = ['count']
```

```python
fig = plt.figure(figsize=(15,7))

for idx, frame in enumerate([cfm_df_plot, dth_df_plot, rcv_df_plot]):
    frame.reset_index()
    if idx == 0:
        plt.plot(frame.index, frame['count'], color='blue', label='Confirmed', marker='o')
    elif idx == 1:
        plt.plot(frame.index, frame['count'], color='red', label='Death')
    elif idx == 2:
        plt.plot(frame.index, frame['count'], color='green', label='Recovered')

plt.legend(loc="upper left")
plt.xlabel("Dates")
plt.ylabel("Count of cases")
plt.figure(figsize=(15,9))
plt.show()
```
